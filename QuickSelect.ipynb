```python
import random
import itertools
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def randomquickselect(alist,times, i):
   comp = [0]
   randomquickselecthelper(alist,0,len(alist)-1, i,comp)
   times.append(comp[0])
```

```python
def randomquickselecthelper( alist, first, last, i, comp ):
    if first == last:
        return alist[first]
    med = randompartition( alist, first, last )
    tmp = comp.pop()
    comp.append( tmp+med[1] )
    
    k = med[0] - first + 1
    if i == k:
        return alist[med[0]], 
    elif i < k:
        return randomquickselecthelper( alist, first, med[0]-1, i, comp )
    else:
        return randomquickselecthelper( alist, med[0]+1, last, i-k, comp )
```

```python
def randompartition(alist,first,last):
  
   indxpiv = random.randint(first,last)
   temp = alist[indxpiv]
   alist[indxpiv] = alist[first]
   alist[first] = temp
   
   pivotvalue = alist[first]
   compspart = 0

   leftmark = first+1
   rightmark = last

   done = False
   while not done:

       while leftmark <= rightmark and alist[leftmark] <= pivotvalue:
           leftmark = leftmark + 1
           compspart = compspart + 1

       while alist[rightmark] >= pivotvalue and rightmark >= leftmark:
           rightmark = rightmark -1
           compspart = compspart + 1

       if rightmark < leftmark:
           done = True
       else:
           temp = alist[leftmark]
           alist[leftmark] = alist[rightmark]
           alist[rightmark] = temp

   temp = alist[first]
   alist[first] = alist[rightmark]
   alist[rightmark] = temp


   return rightmark,compspart
```

```python
def randomquickselecttimedistrib(s,r):
    
    # Create an array of 1 .. n  
    n=s
    runs=r
    x = []
    for i in range(1,n+1):
        x.append(n+1-i)
    
    random.shuffle(x)
    
     # Run quicksort for each permutation
    tlist =[]
    for p in range(1,runs+1):
          y = list(x)
          randomquickselect(y,tlist, random.randint(1,n))
          
        
    plt.hist(tlist)
    plt.title("Number of comparison of Quicksort for all permutaions")
    plt.xlabel("Comparisons of elements")
    plt.ylabel("Frequency")
    plt.show()
    
    print("n",n)
    print("runs",runs)
    print("mean",np.mean(tlist))
    print("stdev",np.std(tlist))
    #print("mean",statistics.mean(tlist))
    #print("stdev",statistics.stdev(tlist))
    print("min",min(tlist))
    print("max",max(tlist))
```
